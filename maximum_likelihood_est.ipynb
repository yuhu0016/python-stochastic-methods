### Part A:
Your child has been randomly selected for Type I diabetes screening, using a highly accurate new test that boasts of a false positive rate of 1% and a false negative rate of 0%. ### 1A Answer 

Should your child test positive, what is the probability that he/she has Type 1 diabetes?

Here we have two mutually excusive outcomes: your child does or does not have Type 1 diabetes.  So Bayes Theorem applies. 
 

Let's set P(A) as the probability that your child has Type 1 diabetes <br />
Let's set P(B) as the probability of a positive test for Type 1 diabetes <br />

We know Bayes Theorem states -

$$ P(A \mid B) = \frac{P(B \mid A) \, P(A)}{P(B)} $$ 

$$ P(A\mid B)=\frac { P(B \mid A)\, P(A) }{ P(B \mid A)P(A) + P(B \mid NA)P(NA) } $$

with P(NA) being the the probability of not A, or the probability of not having Type 1 diabetes.

We know that 0.228% of children have Type 1 diabetes (a priori), so P(A) = 0.00228 <br />

P(NA) = 0.99772 [1-0.00228 = 99.772%, percentage of not having Type 1 diabetes]<br />
P(B | A) = 1 [false negative rate is 0%, so the test will never be negative if the child has diabetes.  Therefore if the child has Type 1 diabetes, the test is positive 100% of the time]<br />
P(B | NA) = .01  [The false positive is 1%. Meaning the test is positive 1% of the time when a child does not have diabetes]<br />

Plugging this into Bayes Theorem we get the probability of my child having Type 1 diabetes given a positive test.

$$ P(A\mid B)=\frac { 1 x 0.00228  }{ 1  x  0.00228 + .01  x  0.99772 } $$

```python
((1*0.00228) / ((0.00228*1)+(0.99772*0.01)))
```

Should you be concerned enough to ask for further testing or treatment for your child?

Even with a positive test, the probability that my child has Type 1 diabetes is only 18.6%  So I would feel concerned enough to ask for additional testing since it is still a low probability. 

Later, you read online that Type I diabetes is 6 times more prevalent in prematurely born children. If this statistic is true, what is the probability that your child, who is prematurely born, has Type I diabetes?

Given the new information, should you be concerned enough to ask for further testing or treatment for your child?<br />

Given this new information, we will need to adjust the a priori, P(A) and also the P(NA) and recalculate. 

We now have -

P(A) = 0.01368 [the new a priori is 6x greater than the old P(A), therefore 1.368% of children have Type 1 diabetes]<br />
P(NA) = 0.98632 [1-1.368% = 98.632%, percentage of children born prematurely do not have Type 1 diabetes]<br />

P(B | A) = 1 [child has Type 1 diabetes, the test is positive 100% of the time]<br />
P(B | NA) = .01  [FALSE POSITIVE - child does not have Type 1 diabetes, the test is positive 1% of the time]<br />

Plugging these into Bayes Theorem we get:

$$ P(A\mid B)=\frac { 1 x 0.01368  }{ 1  x  0.01368 + .01  x  0.98632 } $$



```python
((1*0.01368) / ((0.01368*1)+(0.98632*0.01)))
```

## Problem 2: Maximum Likelihood Estimates

**Coding required for Part B only**

Suppose you observe the following data set $\mathbf{x}^{(0)} = (0.5, 2.5), \mathbf{x}^{(1)} = (3.2, 1.3), \mathbf{x}^{(2)} = (2.72, 5.84), \mathbf{x}^{(3)}= (10.047, 0.354)$. By convention, for any vector $\mathbf{x}$, we will denote the first component of $\mathbf{x}$ by $x_{1}$ and the second component by $x_{2}$. Suppose that the data is drawn from the same two-dimensional probability distribution with pdf $f_X$, that is, $\mathbf{x}^{(i)} \overset{iid}{\sim} f_X$, where
$$
f_X(\mathbf{x}) =  4\lambda_1^2 x_{1}x_{2} \mathrm{exp} \left\{-\lambda_0 (x^2_{1} + x^2_{2}) \right\}.
$$
You should assume that $\lambda_1, \lambda_0 > 0$ and that $f_X$ is supported on the nonnegative quandrant of $\mathbb{R}^2$ (i.e. $f_X$ is zero when either component is negative). 



### Part 2A Answer: 

What are the values for $\lambda_0$ and $\lambda_1$ that maximizes the likelihood of the observed data? 

**Support your answer with full and rigorous analytic derivations.**


In order for the joint distribution to be a pdf, we will integrate it with support of $x_1$ and $x_2$ and set it equal to 1.  This normalizes the pdf. 


$$ \int_{0}^{\infty} \int_{0}^{\infty} 4 \lambda_1^2 x_1 x_2 e^{-\lambda_0 x_1^2}e^{-\lambda_0 x_2^2} dx_1 dx_2 = 1  $$

$$  \int_{0}^{\infty} 4 \lambda_1^2 x_2 e^{-\lambda_0 x_2^2} \int_{0}^{\infty}  x_1 e^{-\lambda_0 x_1^2} dx_1 dx_2 = 1  $$

Breaking this down into inner and outer functions using chain rule by first integrating the inner integral, we get the following:

$$ \int_{0}^{\infty}  x_1 e^{-\lambda_0 x_1^2} dx_1 $$ becomes $$ \frac{1}{2\lambda_0} $$ 
 

Plugging this into the original equation:

$$  \int_{0}^{\infty} 4 \lambda_1^2 x_2 e^{-\lambda_0 x_2^2} \frac{1}{2\lambda_0} dx_2 = 1  $$

$$   \frac{4 \lambda_1^2}{2\lambda_0}  \int_{0}^{\infty}  x_2 e^{-\lambda_0 x_2^2} dx_2 = 1  $$

Further simplifying we can see that $\lambda_0 = \lambda_1$.  Since the prompt tells us both lambdas are positive they must equal each other

$$   \frac{4 \lambda_1^2}{2\lambda_0} \frac{1}{2\lambda_0} = 1  $$

$$   \frac{\lambda_1^2}{\lambda_0^2} = 1  $$


Using maximum likelihood estimation to find lambda, $\hat{\lambda}_{MLE}$, we use the log likelihood equation below


$$l(\lambda) = \mathrm{log} \prod_{i=1}^n (f_X(x_1^{(i)}, x_2^{(i)}))$$


Taking derivatives we get: 

$$ \frac{d}{d\lambda}\log(f_X(\mathbf{x})) = \frac{2}{\lambda} - (x_1^2 + x_2^2) $$

$$ \frac{d^2}{d\lambda}\log(f_X(\mathbf{x})) = - \frac{2}{\lambda^2} $$

Maximizing this we get:  

$$  \sum_\limits{\mathbf{x}} \frac{2}{\lambda} - (x_1^2 + x_2^2) = 0 $$
$$  \lambda = \frac{2*4}{\sum_\limits{\mathbf{x}} (x_1^2 + x_2^2)} = \frac{8}{.5^2 + 2.5^2 + 3.2^2 + 1.3^2 + 2.72^2 + 5.84^2 + 10.047^2 + 0.354^2} =  0.049688970337392765 $$    

To maximize the probability that the values given belong to the distribution given, we want to set the $\lambda_MLE$, or both $\lambda_0$ and $\lambda_1$ to 0.049688970337392765.

<b>Empirically estimate the mean of the distribution $f_{x_1}$. Estimate, also the SE (standard error) of the estimate.</b>

To empirically estimate the mean of the distribution and its standard error, I utilized non-parametric bootstrap sampling.


```python
# to show plots in the notebook
%matplotlib inline
import math

# resample the data using non-parametric bootstrap to estimate the mean
M_samples=10000
N_points = data.shape[0]
bs_np = np.random.choice(data, size=(M_samples, N_points), replace=True)
sd_mean=np.mean(bs_np, axis=1)
sd_std=np.std(bs_np, axis=1)
plt.hist(sd_mean, bins=30, normed=True, alpha=0.5,label="samples");
plt.axvline(data.mean(), 0, 1, color='r', label='Original Data')
plt.legend()
plt.show()

# calculate the estimated mean
print("estimated mean of x1 =", np.mean(sd_mean))

# calculate the standard error of the estimated mean: divide standard deviation of the mean by the sqrt of the sample size.  

print("standard error of the estimate =", np.std(sd_mean) / math.sqrt(10000))
```

<b>Empirically estimate the standard deviation of the distribution $f_{x_2 \mid x_1}$, for $x_1 \in [3.99, 4.01]$. Estimate, also the SE (standard error) of the estimate.</b>

In order to empirically estimate the standard deviation of the distribution and its SE, I will run a non-parametric bootstrap.

```python
# to show plots in the notebook
%matplotlib inline
import math
import pandas as pd
import numpy as np

# pull in the hw data
df = pd.read_table("Homework_1_Data.txt", header=None, sep=',', names=['x1','x2'])

# Subset the data based to include X1 values specified 
df2 = df[(df['x1'] >= 3.99) & (df['x1'] <= 4.01)]

# make X2 subset 
data = df2['x2']

# resample the data using non-parametric bootstrap to estimate the mean
M_samples=10000
N_points = data.shape[0]
bs_np = np.random.choice(data, size=(M_samples, N_points), replace=True)
sd_mean=np.mean(bs_np, axis=1)
sd_std=np.std(bs_np, axis=1)
plt.hist(sd_mean, bins=30, normed=True, alpha=0.5,label="samples");
plt.axvline(data.mean(), 0, 1, color='r', label='Original Data')
plt.legend()
plt.show()

# calculate the standard deviation  
print("estimated standard deviation of x2 given x1 =", np.mean(sd_std))

# calculate the standard error of the standard deviation  
se_of_sd = (np.std(sd_std) / math.sqrt(10000))
print("standard error of the standard deviation =", se_of_sd)
```
